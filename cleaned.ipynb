```python
menu['Date'] = menu['Week'].dt.date
sales['Date'] = sales['timestamp'].dt.date
```

```python
merged = pd.merge(sales, menu, on="Date", how="left")

merged.head(10)

#len(merged)
```

```python
unique_items = merged['Item'].unique()

print(unique_items)
```

```python
# List of items to include
item_list = [
    'Soup - Non-Dickinson Student or Non-Meal Plan',
    'Soup - Dickinson Student w. Meal Plan',
    'Salad - Dickinson Student w. Meal Plan',
    'Salad - Non -Dickinson Student or Non-Meal Plan',
    'Soup Dickinson Staff paying with Declining Balance',
    '1/2 Soup - 1/2 Salad - Non Dickinson Student or Non-Meal Plan',
    '1/2 soup 1/2 salad Dickinson staff paying with Declining Balance',
    'Salad Dickinson Staff paying with Declining Balance',
    '1/2 Soup - 1/2 Salad - Dickinson Student w. Meal Plan'
]

# Subset 
filtered_merged = merged[merged["Item"].isin(item_list)]

filtered_merged.head()
#len(filtered_merged)
```

```python
soup = [
    'Soup - Non-Dickinson Student or Non-Meal Plan',
    'Soup - Dickinson Student w. Meal Plan',
    'Soup Dickinson Staff paying with Declining Balance']

filtered_merged.loc[filtered_merged["Item"].isin(soup), ["Salad", "Empanada"]] = np.nan

salad = [
    'Salad - Dickinson Student w. Meal Plan',
    'Salad - Non -Dickinson Student or Non-Meal Plan',
    'Salad Dickinson Staff paying with Declining Balance']

filtered_merged.loc[filtered_merged["Item"].isin(salad), ["Soup", "Empanada"]] = np.nan

emp = [
    'Monday', 'Tuesday', 'Thursday', 'Friday']

filtered_merged.loc[filtered_merged["day_name"].isin(emp), ["Empanada"]] = np.nan

filtered_merged.head()
```

```python
# Check using Wed sales
wed = filtered_merged[filtered_merged['day_name'] == "Wednesday"]

print(wed)
```

```python
filtered_merged['timestamp'] = pd.to_datetime(filtered_merged['timestamp'])

filtered_merged['date'] = filtered_merged['timestamp'].dt.date
#filtered_merged['day_of_week'] = filtered_merged['timestamp'].dt.dayofweek

half = [
    '1/2 Soup - 1/2 Salad - Non Dickinson Student or Non-Meal Plan',
    '1/2 soup 1/2 salad Dickinson staff paying with Declining Balance',
    '1/2 Soup - 1/2 Salad - Dickinson Student w. Meal Plan']

def create_count_df(time_start, time_end):
    # Filter the DataFrame based on the time range
    subset_time = filtered_merged[
        (filtered_merged['timestamp'].dt.time >= pd.to_datetime(time_start).time()) &
        (filtered_merged['timestamp'].dt.time < pd.to_datetime(time_end).time())
    ]

    # Count full 
    soup_items = subset_time[subset_time['Item'].isin(soup)]
    salad_items = subset_time[subset_time['Item'].isin(salad)]
    #empanada_items = subset_time[subset_time['Item'].str.contains("Empanada", case=False, na=False)]

    # Filter for half Soup and Salad items (add 0.5 to both)
    half_soup_salad_items = subset_time[subset_time['Item'].isin(half)]

    # Count full Soup, Salad, and Empanada items
    soup_count = soup_items.groupby('date').size().reset_index(name='soup_count')
    salad_count = salad_items.groupby('date').size().reset_index(name='salad_count')
    #empanada_count = empanada_items.groupby('date').size().reset_index(name='empanada_count')

    # Count half Soup and Salad items (each counts as 0.5)
    half_soup_salad_count = half_soup_salad_items.groupby('date').size().reset_index(name='half')

    # Add 0.5 for each half Soup and Salad item to the respective columns
    soup_count['soup_count'] += half_soup_salad_count['half'] * 0.5
    salad_count['salad_count'] += half_soup_salad_count['half'] * 0.5

    # Add full counts for Empanadas on Wednesdays (where day_of_week == 2)
    wednesday_half_soup_salad = half_soup_salad_items[half_soup_salad_items['day_name'] == 'Wednesday']
    empanada_count = wednesday_half_soup_salad.groupby('date').size().reset_index(name='empanada_count')

    # Add the full count to the Empanada count on Wednesdays
    #empanada_count = pd.merge(empanada_count, empanada_count_wednesday, on='date', how='outer')
    #empanada_count['empanada_count'] += empanada_count['empanada_count_wednesday'].fillna(0)

    # Merge all counts together
    result = pd.merge(soup_count, salad_count, on='date', how='outer')
    result = pd.merge(result, empanada_count[['date', 'empanada_count']], on='date', how='outer')

    return result

# Create DataFrames for each time range
sales_11_to_12 = create_count_df("11:00:00", "12:00:00")
sales_12_to_1 = create_count_df("12:00:00", "13:00:00")
sales_1_to_2 = create_count_df("13:00:00", "14:00:00")

# Print the resulting DataFrames
print("11 AM to 12 PM:")
print(sales_11_to_12)
print("\n12 PM to 1 PM:")
print(sales_12_to_1)
print("\n1 PM to 2 PM:")
print(sales_1_to_2)
```

```python
# Ensure 'timestamp' column is in datetime format
filtered_merged['timestamp'] = pd.to_datetime(filtered_merged['timestamp'])

# Extract 'date' and 'hour' from 'timestamp' in the larger DataFrame
filtered_merged['date'] = filtered_merged['timestamp'].dt.date
filtered_merged['hour'] = filtered_merged['timestamp'].dt.hour

# Modify the time-specific DataFrames to also have 'date' and 'hour' columns
sales_11_to_12['hour'] = 11
sales_12_to_1['hour'] = 12
sales_1_to_2['hour'] = 13

# Modify the column names in the time-specific DataFrames to reflect the time ranges
sales_11_to_12.rename(columns={'soup_count': 'soup_count_11_12', 
                               'salad_count': 'salad_count_11_12', 
                               'empanada_count': 'empanada_count_11_12'}, inplace=True)

sales_12_to_1.rename(columns={'soup_count': 'soup_count_12_1', 
                              'salad_count': 'salad_count_12_1', 
                              'empanada_count': 'empanada_count_12_1'}, inplace=True)

sales_1_to_2.rename(columns={'soup_count': 'soup_count_1_2', 
                              'salad_count': 'salad_count_1_2', 
                              'empanada_count': 'empanada_count_1_2'}, inplace=True)

# Now merge the time-specific DataFrames based on both 'date' and 'hour'
merged_11_12 = pd.merge(filtered_merged, sales_11_to_12, on=['date', 'hour'], how='left')
merged_12_1 = pd.merge(merged_11_12, sales_12_to_1, on=['date', 'hour'], how='left')
final_merged = pd.merge(merged_12_1, sales_1_to_2, on=['date', 'hour'], how='left')

# Display the final merged DataFrame
print(final_merged)


```
