```python
def create_count_df():
    # Count full
    soup_items = filtered_merged[filtered_merged['Item'].isin(soup)]
    salad_items = filtered_merged[filtered_merged['Item'].isin(salad)]

    # Filter for half Soup and Salad items (add 0.5 to both)
    half_soup_salad_items = filtered_merged[filtered_merged['Item'].isin(half)]

    # Filter for Empanada items based on year and day
    empanada_items = filtered_merged[
        (filtered_merged['Item'].str.contains("Empanada", case=False, na=False)) &
        (
            ((filtered_merged['year'] == 2023) & (filtered_merged['day_name'] == 'Wednesday')) |
            ((filtered_merged['year'] == 2024) & (filtered_merged['day_name'].isin(['Tuesday', 'Thursday'])))
        )
    ]

    # Count full Soup, Salad, and Empanada items
    soup_count = soup_items.groupby('date').size().reset_index(name='soup_count')
    salad_count = salad_items.groupby('date').size().reset_index(name='salad_count')
    empanada_count = empanada_items.groupby('date').size().reset_index(name='empanada_count')

    # Count half Soup and Salad items (each counts as 0.5)
    half_soup_salad_count = half_soup_salad_items.groupby('date').size().reset_index(name='half')

    # Add 0.5 for each half Soup and Salad item to the respective columns
    soup_count = pd.merge(soup_count, half_soup_salad_count, on='date', how='outer')
    salad_count = pd.merge(salad_count, half_soup_salad_count, on='date', how='outer')

    soup_count['soup_count'] = soup_count['soup_count'].fillna(0) + soup_count['half'].fillna(0) * 0.5
    salad_count['salad_count'] = salad_count['salad_count'].fillna(0) + salad_count['half'].fillna(0) * 0.5

    # Keep soup, salad, and empanada counts separate
    soup_count = soup_count[['date', 'soup_count']]
    salad_count = salad_count[['date', 'salad_count']]
    empanada_count = empanada_count[['date', 'empanada_count']]

    # Merge all counts together without summing them
    result = pd.merge(soup_count, salad_count, on='date', how='outer')
    result = pd.merge(result, empanada_count, on='date', how='outer')

    return result

# Create the summary DataFrame for all dates
sales_summary = create_count_df()

# Print the resulting DataFrame
print("Daily Sales Summary:")
print(sales_summary)

# Merge the counts with the original data without summing
merged = pd.merge(filtered_merged, sales_summary, on=['date'], how='left')

# Assuming `cost` has additional information for merging; adjust as needed
merged1 = pd.merge(merged, cost, on='Soup', how='left')

print(merged1)

# Save the resulting DataFrame to an Excel file
merged.to_excel("sales_menu_cleaned1.xlsx", index=False, engine='openpyxl')
```
